```python
import random
#Step 1: Creating this colab

#Step 2: Creating a class and display board
#Creating a class for a TicTacToe game
class TTT:
  def __init__(self):
    self.board = ['1', '2', '3', '4', '5', '6', '7', '8', '9']
    self.board_original = ['1', '2', '3', '4', '5', '6', '7', '8', '9']
    self.current_winner = None
    self.player_one = "Player_one"
    self.player_two = "Player_two"
    self.whose_turn= 0
    self.played_moves = []
    self.turn_counter = 1
    self.end_game= False
    self.play_again= True

  #New method: display_board
  def display_board(self):
    print(f" {self.board[0]} | {self.board[1]} | {self.board[2]}")
    print("-----------")
    print(f" {self.board[3]} | {self.board[4]} | {self.board[5]}")
    print("-----------")
    print(f" {self.board[6]} | {self.board[7]} | {self.board[8]}")


#Step 3: Player Input and Marker Selection (Making the X's and O's)
  def player_input(self):
    selection = input("Player One, X or O? ")
    print(selection)
    self.player_one = selection
    if selection == "X":
      self.player_two = "O"
    if selection == "O":
      self.player_two = "X"
    print(f"Player One will be {self.player_one} and Player Two will be {self.player_two}.")

#Step 4
#Figuring out who is going first
  def choose_first(self):
    first = random.randint(1, 2)
    self.whose_turn = first
    if self.whose_turn == 1:
      print("Player One, you will be going first!")

    else:
      print("Player Two, you will be going first!")

#Prompting the player who is going first to pick a place on the board
  def player_choice(self):
    ask_position = int(input("Picking from numbers 1 through 9, where would you want that place? "))
#pls try again and make it 1-9
    while (ask_position <1 and ask_position >9) or (ask_position in self.played_moves):
          ask_position = int(input("Please try again! Picking from numbers 1 through 9, where would you want that place? "))
    self.played_moves.append(ask_position)
    return ask_position

#Place the marker wherever on the board the player wants it
  def player_marker(self, choice):
    if self.whose_turn==1:
      the_marker= self.player_one
    elif self.whose_turn==2:
      the_marker= self.player_two

    if choice == 1:
      self.board[0] = the_marker
    elif choice == 2:
      self.board[1] = the_marker
    elif choice == 3:
      self.board[2] = the_marker
    elif choice == 4:
      self.board[3] = the_marker
    elif choice == 5:
      self.board[4] = the_marker
    elif choice == 6:
      self.board[5] = the_marker
    elif choice == 7:
      self.board[6] = the_marker
    elif choice == 8:
      self.board[7] = the_marker
    elif choice == 9:
      self.board[8] = the_marker

#Switches Turn
    if self.whose_turn==1:
      self.whose_turn= self.whose_turn + 1

    elif self.whose_turn==2:
      self.whose_turn= self.whose_turn - 1

#Full Board Check
  def full_board_check(self):
    if self.turn_counter ==9:
      print("You have ended in a draw.")
      self.end_game=True
    else:
      self.turn_counter += 1

#Step 5 (Check Win) #see if the game goes at least 3x first, then check for the selection in (1,2,3),etc.
  def check_win(self):
    if self.turn_counter >= 5:
      if(self.board[0] == self.board[1] and self.board[0] == self.board[2]) or (self.board[0] == self.board[3] and self.board[0] == self.board[6]) or (self.board[3] == self.board[4] and self.board[3] == self.board[5]) or (self.board[6] == self.board[7] and self.board[6] == self.board[8]) or (self.board[1] == self.board[4] and self.board[1] == self.board[7]) or (self.board[2] == self.board[5] and self.board[2] == self.board[8]) or (self.board[0] == self.board[4] and self.board[0] == self.board[8]) or (self.board[2] == self.board[4] and self.board[2] == self.board[6]):

        self.end_game=True
        if self.whose_turn==2:
          print ("Player One has won!")
        elif self.whose_turn==1:
          print ("Player Two has won!")

#Step 6: End, Replay and reset the game
  def play_game(self):
    self.player_input()
    self.choose_first()
    while self.play_again == True:
      self.display_board()
      choice=self.player_choice()
      self.player_marker(choice)
      self.full_board_check()
      self.check_win()
      if self.end_game ==True:
        self.display_board()
        ans=input("Would you like to play again? y/n")
        if ans=="y" or ans== "Y":
          self.play_again=True
          self.board=self.board_original
          self.played_moves.clear()
          self.end_game=False
          self.turn_counter=1
        else:
          self.play_again=False

ttt = TTT()
ttt.play_game()
```
